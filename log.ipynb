```python
"""
log.py

Personal reading log for the user.
Marking a book as "read" pushes it onto the recently_read Stack.
Users can also rate (1–5) and/or write a review for any logged book

Functions
mark_as_read(bst, recently_read)
rate_and_review(bst)
view_recently_read(recently_read)
run_log_menu(bst, recently_read)
"""

from bst import BST
from stack import Stack
from book import Book

def mark_as_read(bst: BST, recently_read: Stack):
  """
  Look up a book by ISBN, prompt for optional rating/review
  then push it onto the recently_read Stack
  """
  isbn = input("  Enter ISBN of the book you read: ").strip()
  book = bst.search(isbn)
  if book == None:
    print("  No book found with ISBN '" + isbn + "'. Make sure it's in the database.")
    return

  print(f"\n  Logging: \"{book.title}\" by {book.author}")
  prompt_rating_review(book)
  recently_read.push(book)
  print(f"  ✓ \"{book.title}\" added to your reading log.")


def rate_and_review(bst: BST):
  """
  Update the rating and/or review for a book (whether or not it has been marked as read
  """
  isbn = input("  Enter ISBN of the book to rate/review: ").strip()
  book = bst.search(isbn)
  if not book:
    print(f"  No book found with ISBN {isbn}.")
    return

  print(f"\n  Book: \"{book.title}\" by {book.author}")
  if book.rating:
    current_rating = f"{book.rating}/5"
  else:
    current_rating = "None"
  if book.review:
    current_review = book.review
  else:
    current_review = "None"
  print(f"  Current rating : {current_rating}")
  print(f"  Current review : {current_review}")
  prompt_rating_review(book)
  print("  ✓ Rating/review updated.")


def prompt_rating_review(book):
  """ask the user for an optional rating and review, then save to book."""
  # Rating
  while True:
    raw = input("  Rate this book (1–5), or press Enter to skip: ").strip()
    if raw == "":
      break
    if raw.isdigit():
      if 1 <= int(raw) <= 5:
        book.rating = int(raw)
      break
    print("  Please enter a number between 1 and 5, or press Enter to skip.")

  # Review
  review = input("  Write a short review (or press Enter to skip): ").strip()
  if review:
    book.review = review


def view_recently_read(recently_read: Stack):
  """Print the recently-read Stack contents (most recent first)."""
  books = recently_read.to_list()
  if not books:
    print("\n  Your reading log is empty.")
    return
  print(f"\n  Recently Read — {len(books)} book(s), most recent first:")
  print("  " + "-" * 50)
  i = 1
  for book in books:
    if book.rating != None:
      stars = str(book.rating) + "/5"
    else:
      stars = "Not rated"
    print(f"  {i}. \"{book.title}\" by {book.author}  [{stars}]")
    if book.review:
      print(f"      Review: {book.review}")
    i += 1
  print("  " + "-" * 50)


def add_book_to_database(bst: BST):
    """
    Allow the user to manually add a new book to the BST database
    """
    print("\n  — Add a New Book to the Database —")
    isbn = input("  ISBN   : ").strip()
    if not isbn:
      print("  ISBN cannot be empty")
    if bst.search(isbn):
      print(f"  A book with ISBN '{isbn}' already exists.")
      return
    title  = input("  Title  : ").strip()
    author = input("  Author : ").strip()
    year   = input("  Year   : ").strip()

    book = Book(isbn, title, author, year)
    bst.insert(book)
    print(f"    \"{title}\" added to the database.")

def run_log_menu(bst: BST, recently_read: Stack):
  while True:
    print("\n╔══════════════════════════════╗")
    print("║       MY READING LOG         ║")
    print("╠══════════════════════════════╣")
    print("║  1. Mark a book as read      ║")
    print("║  2. Rate / review a book     ║")
    print("║  3. View recently read       ║")
    print("║  4. Add new book to database ║")
    print("║  5. Back to main menu        ║")
    print("╚══════════════════════════════╝")
    choice = input("  Choice: ").strip()

    if choice == "1":
        mark_as_read(bst, recently_read)
    elif choice == "2":
        rate_and_review(bst)
    elif choice == "3":
        view_recently_read(recently_read)
    elif choice == "4":
        add_book_to_database(bst)
    elif choice == "5":
        break
    else:
        print("  Invalid choice. Please try again.")
```
