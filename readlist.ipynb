```python
"""
readlist.py
"To-Read List" — a Queue of books the user wants to read next.
Books are added to the rear of the queue and removed from the front (FIFO),
so the first book added is the first one suggested.

Menu options
Add a book to the to-read list
View the full to-read list
Mark next book as read (dequeue front → push onto recently_read)
Remove a specific book from the list
Back to main menu
"""

from bst import BST
from bookqueue import Queue
from stack import Stack
from log import _prompt_rating_review


def add_to_readlist(bst: BST, readlist: Queue):
  """Find a book by ISBN and enqueue it"""
  isbn = input("  Enter ISBN of the book to add: ").strip()
  book = bst.search(isbn)
  if not book:
    print(f"  No book found with ISBN '{isbn}'.")
    return
  if readlist.contains(isbn):
    print(f"  \"{book.title}\" is already in your to-read list.")
    return
  readlist.enqueue(book)
  print(f"  ✓ \"{book.title}\" added to your to-read list.")


def view_readlist(readlist: Queue):
  """Print the full to-read queue (front = next to read)."""
  print("\n" + str(readlist))


def read_next(readlist: Queue, recently_read: Stack):
  """
  Dequeue the front book (next to read), optionally rate/review it,
  then push it onto the recently_read Stack
  """
  book = readlist.dequeue()
  if book is None:
    print("\n  Your to-read list is empty.")
    return

  print(f"\n  Next up: \"{book.title}\" by {book.author}")
  log_it = input("  Mark this as read and log it? (y/n): ").strip().lower()
  if log_it == "y":
    _prompt_rating_review(book)
    recently_read.push(book)
    print(f"  ✓ \"{book.title}\" moved to your reading log.")
  else:
    # Put it back at the front if user changed their mind
    temp = readlist.to_list()
    # Clear the queue
    while not readlist.is_empty():
      readlist.dequeue()
    # Re-insert book at front, then the rest
    readlist.enqueue(book)
    for b in temp:
      readlist.enqueue(b)
    print("  Okay — book returned to the front of your list.")


def remove_from_readlist(readlist: Queue):
  """Remove a specific book by ISBN"""
  isbn = input("  Enter ISBN of the book to remove: ").strip()
  removed = readlist.remove(isbn)
  if removed:
    print(f"  ✓ \"{removed.title}\" removed from your to-read list.")
  else:
    print(f"  No book with ISBN {isbn} found in your to-read list.")



def run_readlist_menu(bst: BST, readlist: Queue, recently_read: Stack):
  while True:
    print("\n╔══════════════════════════════╗")
    print("║        TO-READ LIST          ║")
    print("╠══════════════════════════════╣")
    print("║  1. Add book to list         ║")
    print("║  2. View to-read list        ║")
    print("║  3. Start reading next book  ║")
    print("║  4. Remove a book from list  ║")
    print("║  5. Back to main menu        ║")
    print("╚══════════════════════════════╝")
    choice = input("  Choice: ").strip()

    if choice == "1":
      add_to_readlist(bst, readlist)
    elif choice == "2":
      view_readlist(readlist)
    elif choice == "3":
      read_next(readlist, recently_read)
    elif choice == "4":
      remove_from_readlist(readlist)
    elif choice == "5":
      break
    else:
      print("  Invalid choice. Please try again.")
```
