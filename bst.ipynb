```python
'''BST.py
binary search tree storing books and sorting them by ISBN

insert(book)
search(isbn)
delete(isbn)
inorder()
search_by_title()
search_by_author()

'''


from book import Book

class BSTNode:
  def __init__(self, book: Book):
    self.book = book
    self.left = None
    self.right = None
class BST:
  def __init__(self):
    self.root = None


  '''insert method'''
  def insert(self, book: Book):
    self.root = self._insert(self.root, book)

  def _insert(self, node, book):
    if node is None:
      return BSTNode(book)
    if book.isbn < node.book.isbn:
      node.left = self._insert(node.left, book)
    elif book.isbn > node.book.isbn:
      node.right = self._insert(node.right, book)
    else:
      node.book = book
    return node


  '''search by exact ISBN'''
  def search(self, isbn: str):
    current = self.root
    search_isbn = isbn.strip()

    while current is not None:
      if search_isbn == current.book.isbn:
        return current.book
      elif search_isbn < current.book.isbn:
        current = current.left
      else:
        current = current.right

    return None

  '''DELETE method'''

  def delete(self, isbn: str):
    self.root = self._delete(self.root, isbn.strip())

  def _delete(self, node, isbn):
    if node is None:
      return None
    if isbn < node.book.isbn:
      node.left = self._delete(node.left, isbn)
    elif isbn > node.book.isbn:
      node.right = self._delete(node.right, isbn)
    else:
      if node.left is None:
        return node.right
      elif node.right is None:
        return node.left


      successor = self._min_node(node.right)
      node.book = successor.book
      node.right = self._delete(node.right, successor.book.isbn)
    return node

  def _min_node(self, node):
    while node.left is not None:
      node = node.left
    return node


  '''in order traversal'''
  def inorder(self):
    '''return a list of all books sorted by isbn'''
    result = []
    self._inorder(self.root, result)
    return result

  def _inorder(self, node, result):
    if node is None:
      return
    self._inorder(node.left, result)
    result.append(node.book)
    self._inorder(node.right, result)

  '''search by title'''
  def search_by_title(self, keyword: str):
    ''' return all books whose title contains key word, case insenstive'''
    kw = keyword.lower()
    results = []
    all_books = self.inorder()
    for b in all_books:
      if kw in b.title.lower():
        results.append(b)

  def search_by_author(self, keyword: str):
    '''return all books whose author contains insensitve keyword'''
    kw = keyword.lower()
    return [b for b in self.inorder() if kw in b.author.lower()]


```
