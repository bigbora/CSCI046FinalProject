```python
"""
book.py
Defines the Book data class used throughout
Each book is identified by ISBN
"""

class Book:
  def __init__(self, isbn, title, author, year):
    self.isbn = isbn.strip()
    self.title = title.strip()
    self.author = author.strip()
    self.year = year.strip()

    # User-added data
    self.rating = None
    self.review = ""

  def __str__(self):
    if self.rating:
      stars = f"{self.rating}/5 ★"
    else:
      stars = "Not rated"

    lines = [
      f"  Title      : {self.title}",
      f"  Author     : {self.author}",
      f"  Year       : {self.year}",
      f"  ISBN       : {self.isbn}",
      f"  Rating     : {stars}",
    ]
    if self.review:
      lines.append(f"  Review     : {self.review}")
    return "\n".join(lines)
```
