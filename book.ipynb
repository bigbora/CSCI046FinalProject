```python
"""
book.py
-------
Defines the Book data class used throughout the system.
Each book is identified by its ISBN (used as the BST key).
"""


class Book:
  def __init__(self, isbn, title, author, year):
    self.isbn = isbn.strip()
    self.title = title.strip()
    self.author = author.strip()
    self.year = year.strip()

    # User-added data
    self.rating = None      # int 1–5, set when user logs the book
    self.review = ""        # str, set when user logs the book

  def __str__(self):
    stars = f"{self.rating}/5 ★" if self.rating else "Not rated"
    lines = [
      f"  Title      : {self.title}",
      f"  Author     : {self.author}",
      f"  Year       : {self.year}",
      f"  ISBN       : {self.isbn}",
      f"  Rating     : {stars}",
    ]
    if self.review:
      lines.append(f"  Review     : {self.review}")
    return "\n".join(lines)
```
