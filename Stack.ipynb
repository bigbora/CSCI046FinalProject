```python
'''
stack.py
stack using a singly linked list

used for:

search history (most recently marked as searched on top)
recently read (most recently marked as read on top)

supported operations:

push(item)  add item to top
pop()       remove and return top item
peek()      view the top item without returning
is_empty()  true if stack has no items
to_list()   return all items top-to-bottom
'''

class stackNode:
  def __init__(self, data):
    self.data = data
    self.next = None



class Stack:
  def __init__(self, name = "Stack", max_size=50):
    '''parameters
    name: label shown when printing stack contents
    max_size: oldest entries are dropped once stack reaches this size'''

    self._top = None
    self._size = 0
    self.name = name
    self.max_size = max_size

  def push(self, item):
    '''push item (book) on top of stack'''
    node = stackNode(item)
    node.next = self._top
    self._top = node
    self._size += 1

    #trim oldest entry if we are exceeding max_size

    if self._size > self.max_size:
      self._trim()


  def _trim(self):
    '''remove the oldest element to stay within max size'''
    if self._top is None:
      return
    current = self._top
    while current.next and current.next.next:
      current = current.next
    current.next = None
    self._size -= 1

  def pop(self):
    '''remove and return the top item, return None if empty'''
    if self.is_empty():
      return None
    data = self._top.data
    self._top = self._top.next
    self._size -= 1
    return data

  def peek(self):
    '''return top element without removing'''
    if self.is_empty():
      return None
    return self._top.data

  def is_empty(self):
    return (self._top == None)

  def size(self):
    return self._size

  def to_list(self):
    '''return items as a list'''
    result = []
    current = self._top
    while current:
      result.append(current.data)
      current = current.next

    return result

  def __str__(self):
    items = self.to_list()
    if not items:
      return f"[{self.name}] (empty)"
    lines = [f"[{self.name}] — {len(items)} item(s), most recent first:"]

    for i, item in enumerate(items, 1):
      lines.append(f"  {i}. {item}")
    return "\n".join(lines)
```
