```python
'''
stack.py
stack using a singly linked list

used for:
search history (most recently marked as searched on top)
recently read (most recently marked as read on top)
'''

class stackNode:
  def __init__(self, data):
    self.data = data
    self.next = None



class Stack:
  def __init__(self, name = "Stack"):
    '''parameters
    name: label shown when printing stack contents'''

    self.top = None
    self.count = 0
    self.name = name

  def push(self, item):
    '''push item (book) on top of stack'''

    node = stackNode(item)
    node.next = self.top
    self.top = node
    self.count += 1

  def pop(self):
    '''remove and return the top item, return None if empty'''
    if self.is_empty():
      return None
    removed_data = self.top.data
    self.top = self.top.next
    self.count = self.count - 1
    return removed_data

  def peek(self):
    '''return top element without removing'''
    if self.is_empty():
      return None
    return self.top.data

  def is_empty(self):
    return (self.top == None)

  def size(self):
    return self.count

  def to_list(self):
    '''return items as a list'''
    result = []
    current = self.top
    while current:
      result.append(current.data)
      current = current.next

    return result

  def __str__(self):
    '''string building for the stack display'''
    if self.is_empty():
      return self.name + " is empty."

    output = f"[ {self.name} ] Content (Top First):\n"
    current = self.top
    index = 1

    while current is not None:
      output += "  " + str(index) + ". " + str(current.data) + "\n"
      current = current.next
      index += 1

    return output
```
