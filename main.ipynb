```python
"""
main.py
Entry point for the Book Search & Rating System.
Initialises all data structures and loads book database

Data structures in use
BST   (bst.py)
Stack (stack.py)
Queue (queue.py)
"""

from bst import BST
from stack import Stack
from bookqueue import Queue
from load_books import load_from_csv
from search import run_search_menu
from log import run_log_menu
from readlist import run_readlist_menu

CSV_FILE = "booksclean.csv"   # swap for your own dataset path


def print_banner():
  print("=" * 50)
  print("     📚  BOOK SEARCH & RATING SYSTEM  📚")
  print("=" * 50)


def print_main_menu():
  print("\n╔══════════════════════════════╗")
  print("║          MAIN MENU           ║")
  print("╠══════════════════════════════╣")
  print("║  1. Search / Browse books    ║")
  print("║  2. My reading log           ║")
  print("║  3. My to-read list          ║")
  print("║  4. View search history      ║")
  print("║  5. Quit                     ║")
  print("╚══════════════════════════════╝")


def view_search_history(search_history: Stack):
  """Print the search history stack (most recent first)."""
  queries = search_history.to_list()
  if not queries:
    print("\n  No searches recorded yet.")
    return
  print(f"\n  Search History — {len(queries)} query/queries, most recent first:")
  i = 1
  for q in queries:
    print(f"    {i}. {q}")
    i += 1

def main():
  book_db       = BST()           # All books, sorted by ISBN
  search_history = Stack(name="Search History")
  recently_read  = Stack(name="Recently Read")
  readlist       = Queue(name="To-Read List")

  # ── Load book database from CSV
  print_banner()
  count = load_from_csv(book_db, CSV_FILE)
  if count:
    print(f"  Loaded {count} books from '{CSV_FILE}'.")
  else:
    print(f"  Warning: No books loaded. Check that '{CSV_FILE}' exists.")

  while True:
    print_main_menu()
    choice = input("  Choice: ").strip()

    if choice == "1":
      run_search_menu(book_db, search_history)

    elif choice == "2":
      run_log_menu(book_db, recently_read)

    elif choice == "3":
      run_readlist_menu(book_db, readlist, recently_read)

    elif choice == "4":
      view_search_history(search_history)

    elif choice == "5":
      print("\n  Goodbye! Happy reading. 📖\n")
      break

    else:
      print("  Invalid choice. Please enter 1–5.")


if __name__ == "__main__":
  main()
```
