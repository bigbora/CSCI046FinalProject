```python
"""
load_books.py

Reads a CSV file and inserts each row as a Book into the BST.

Expected CSV columns (order matters, header row required):
  isbn, title, author, year

Usage
  from load_books import load_from_csv
  load_from_csv(bst, "booksclean.csv")
"""

import csv
from bst import BST
from book import Book


def load_from_csv(bst, filepath):
  """
  Load books from a CSV file into BST
  bst: The BST to populate.
  filepath : Path to the CSV

  Returns:
  Number of books successfully loaded.
  """
  loaded = 0

  try:
    file = open(filepath, "r", encoding="utf-8")

    # skip header line
    header = file.readline()

    # Process the rest of the lines
    for line in file:
      # Strip the newline and split by comma
      line = line.strip()
      if line == "":
        continue #Skipping empty lines

      parts = line.split(",")

      # index 0: isbn, 1: title, 2: author, 3: year
      if len(parts) >= 4:
        isbn = parts[0].strip()
        title = parts[1].strip()
        author = parts[2].strip()
        year = parts[3].strip()


        # Only add if we have the essential
        if isbn != "" and title != "":
          new_book = Book(isbn, title, author, year)
          bst.insert(new_book)
          loaded += 1

    file.close()

  except FileNotFoundError:
    print("Error: The file '" + filepath + "' was not found.")
  except Exception as e:
    print("An error occurred while loading books.")

  return loaded
```
