```python
"""
load_books.py
-------------
Reads a CSV file and inserts each row as a Book into the BST.

Expected CSV columns (order matters, header row required):
  isbn, title, author, genre, year, description

Usage
-----
  from load_books import load_from_csv
  load_from_csv(bst, "sample_books.csv")
"""

import csv
from bst import BST
from book import Book


def load_from_csv(bst, filepath):
  """
  Load books from a CSV file into the BST.

  Parameters
  ----------
  bst      : The BST instance to populate.
  filepath : Path to the CSV file.

  Returns
  -------
  Number of books successfully loaded.
  """
  loaded = 0

  try:
    file = open(filepath, "r", encoding="utf-8")

    # Skip the header row (the first line)
    header = file.readline()

    # Process the rest of the lines
    for line in file:
      # Strip the newline character and split by comma
      line = line.strip()
      if line == "":
        continue # Skip empty lines

      parts = line.split(",")

      # index 0: isbn, 1: title, 2: author, 3: genre, 4: year, 5: description
      if len(parts) >= 5:
        isbn = parts[0].strip()
        title = parts[1].strip()
        author = parts[2].strip()
        genre = parts[3].strip()
        year = parts[4].strip()

        # Description might be optional or contain extra commas
        description = ""
        if len(parts) > 5:
          description = parts[5].strip()

        # Only add if we have the essentials
        if isbn != "" and title != "":
          new_book = Book(isbn, title, author, genre, year, description)
          bst.insert(new_book)
          loaded += 1

    file.close()

  except FileNotFoundError:
    print("Error: The file '" + filepath + "' was not found.")
  except Exception as e:
    print("An error occurred while loading books.")

  return loaded
```
