```python
from bst import BST
from stack import Stack

def search_by_isbn(bst, isbn):
  """Exact ISBN lookup via BST"""
  return bst.search(isbn)

def search_by_title(bst, keyword):
  """Return books whose title contains keyword"""
  return bst.search_by_title(keyword)

def search_by_author(bst, author_name):
  """Return books whose author contains keyword"""
  return bst.search_by_author(author_name)

def list_all_books(bst):
  """Return all books from the BST inorder traversal"""
  return bst.inorder()

def print_results(books, query_label):
  """Print the results list"""
  if len(books) == 0:
    print(f"\n  No books found for: {query_label}")
  else:
    print("\n  Found " + str(len(books)) + " book(s) for " + query_label + ":")
    print("  --------------------------------------------------")
    for book in books:
      print(str(book))
      print("  --------------------------------------------------")

def run_search_menu(bst, search_history):
  """
  Interactive menu for searching.
  """
  keep_running = True
  while keep_running:
    print("\n==============================")
    print("        SEARCH BOOKS          ")
    print("==============================")
    print("  1. Search by ISBN")
    print("  2. Search by Title")
    print("  3. Search by Author")
    print("  4. List all books")
    print("  5. Back to main menu")
    print("==============================")

    choice = input("  Choice: ")

    if choice == "1":
      isbn = input("  Enter ISBN: ")
      book = search_by_isbn(bst, isbn)
      if book != None:
        print("\n" + str(book))
        search_history.push("ISBN: " + isbn)
      else:
        print("  No book found with ISBN: " + isbn)

    elif choice == "2":
      kw = input("  Enter title keyword: ")
      results = search_by_title(bst, kw)
      print_results(results, kw)
      if len(results) > 0:
        search_history.push("Title: " + kw)

    elif choice == "3":
      kw = input("  Enter author name: ")
      results = search_by_author(bst, kw)
      print_results(results, kw)
      if len(results) > 0:
        search_history.push("Author: " + kw)

    elif choice == "4":
      all_books = list_all_books(bst)
      print_results(all_books, "All Books")

    elif choice == "5":
      keep_running = False

    else:
      print("  Invalid choice. Please try again.")
```
