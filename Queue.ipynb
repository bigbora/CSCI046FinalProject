```python
"""
queue.py
Queue using a singly linked list
for the "To-Read List" (readlist).

Books are enqueued at the rear and dequeued from the front FIFO)

Supported operations:
enqueue(item)
dequeue()
peek()
contains(isbn)
remove(isbn)
is_empty()
to_list()
"""


class QueueNode:
  def __init__(self, data):
    self.data = data
    self.next = None


class Queue:
  def __init__(self, name="Queue"):
    self.front = None
    self.rear = None
    self.count = 0
    self.name = name

  def enqueue(self, item):
    """add item to rear of queue"""
    node = QueueNode(item)
    if self.rear is None:
      self.front = self.rear = node
    else:
      self.rear.next = node
      self.rear = node
    self.count += 1

  def dequeue(self):
    """remove and return front item"""
    if self.is_empty():
      return None
    data = self.front.data
    self.front = self.front.next
    if self.front is None:
      self.rear = None
    self.count -= 1
    return data

  def peek(self):
    """return front item without removing"""
    if self.is_empty():
      return None
    return self.front.data

  def contains(self, isbn):
    """searches linearly through the nodes"""
    current = self.front
    while current:
      if current.data.isbn == isbn:
        return True
      current = current.next
    return False

  def remove(self, isbn):
    """removing node by isbn"""
    prev = None
    current = self.front
    while current:
      if current.data.isbn == isbn:
        if prev is None:
          self.front = current.next
        else:
          prev.next = current.next

        if current.next is None:
          self.rear = prev
        self.count -= 1
        return current.data
      prev = current
      current = current.next
    return None

  def is_empty(self):
    return self.front is None

  def count(self):
    return self.count

  def to_list(self):
      """Return all items as a list, starting from the front first"""
      result = []
      current = self.front
      while current:
        result.append(current.data)
        current = current.next
      return result

  def __str__(self):
    if self.is_empty():
      return f"[ {self.name} ] is empty."

    output = f"[ {self.name} ] — {str(self.count)} books:\n"
    current = self.front
    index = 1

    while current is not None:
      book = current.data
      output += f"  {str(index)}.  {book.title} by {book.author} \n"
      current = current.next
      index += 1

    return output
```
