```python
"""
queue.py
--------
Queue using a singly linked list.
for the "To-Read List" (readlist).

Books are enqueued at the rear and dequeued from the front,
preserving the order in which the user added them (FIFO).

Supported operations:
  - enqueue(item)  : Add item to the rear of the queue
  - dequeue()      : Remove and return the front item
  - peek()         : View the front item without removing
  - contains(isbn) : Check if a book ISBN is already in the queue
  - remove(isbn)   : Remove a specific book by ISBN (mid-queue removal)
  - is_empty()     : True if queue has no items
  - to_list()      : Return all items front-to-rear (non-destructive)
"""


class _QueueNode:
  def __init__(self, data):
      self.data = data
      self.next = None


class Queue:
  def __init__(self, name="Queue"):
      self._front = None
      self._rear = None
      self._size = 0
      self.name = name

  def enqueue(self, item):
      """add item to rear of queue"""
      node = _QueueNode(item)
      if self._rear is None:
          self._front = self._rear = node
      else:
          self._rear.next = node
          self._rear = node
      self._size += 1

  def dequeue(self):
      """Remove and return the front item. Returns None if empty."""
      if self.is_empty():
          return None
      data = self._front.data
      self._front = self._front.next
      if self._front is None:
          self._rear = None
      self._size -= 1
      return data

  def peek(self):
      """Return the front item without removing it. Returns None if empty."""
      if self.is_empty():
          return None
      return self._front.data

  def contains(self, isbn: str):
      """Return True if a book with the given ISBN is already in the queue."""
      current = self._front
      while current:
          if current.data.isbn == isbn:
              return True
          current = current.next
      return False

  def remove(self, isbn: str):
      """
      Remove a book by ISBN from anywhere in the queue.
      Returns the removed Book, or None if not found.
      """
      prev = None
      current = self._front
      while current:
          if current.data.isbn == isbn:
              if prev is None:
                  self._front = current.next
              else:
                  prev.next = current.next
              if current.next is None:
                  self._rear = prev
              self._size -= 1
              return current.data
          prev = current
          current = current.next
      return None

  def is_empty(self):
      return self._front is None

  def size(self):
      return self._size

  def to_list(self):
      """Return all items as a list, front first (non-destructive)."""
      result = []
      current = self._front
      while current:
          result.append(current.data)
          current = current.next
      return result

  def __str__(self):
      items = self.to_list()
      if not items:
          return f"[{self.name}] (empty)"
      lines = [f"[{self.name}] — {len(items)} book(s), next up first:"]
      for i, book in enumerate(items, 1):
          lines.append(f"  {i}. \"{book.title}\" by {book.author}")
      return "\n".join(lines)
```
